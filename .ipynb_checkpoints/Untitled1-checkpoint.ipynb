### Fine-Tuning & Prepocessing the datasets

```python
#good
import xml.etree.ElementTree as ET
from benchmark_reader import Benchmark
from benchmark_reader import select_files
from sklearn.model_selection import train_test_split
from transformers import T5Tokenizer, T5ForConditionalGeneration
import torch
from torch.nn.utils.rnn import pad_sequence

tokenizer = T5Tokenizer.from_pretrained('t5-base')
model = T5ForConditionalGeneration.from_pretrained('t5-base')

# using benchmark_reader.py for extracting data from the dataset
b = Benchmark()
files = select_files('en/train')
b.fill_benchmark(files)

triples_texts = [] 
predicate_train = []
max_triples = 0
for i in range(b.entry_count()):
    #print('====')
    entry = b.entries[i]
    
    length = entry.size
    #print(type(length))
    if length == '1':
        category = entry.category
        data = entry.list_triples()
        texts = entry.lexs
        triple = []
        text = []

        for RDF in range(len(entry.list_triples())):
            #dividing in the Triples into 3 strings (Subject, Predicate and Object)
            input_string = data[RDF]

            strings = input_string.split(' | ')
            string1 = strings[0]
            string2 = strings[1]
            string3 = strings[2]

            # Removing unwanted symbols
            if '_' in string1:
                string1 = string1.replace('_', ' ')

            if '_' in string2:
                string2 = string2.replace('_', ' ')

            if '"' in string3:
                string3 = string3.replace('"', '')

            MR = "__subject__ " + string1 + " __predicate__ " + string2 + " __object__ " + string3
            triple.append(MR)
            predicate_train.append(string2)
            max_triples= max_triples+1

        for lex in range(len(texts)):
            text.append(texts[lex].lex)
        
        triples_texts.append((triple, [text[0]]))

train_data = triples_texts

# using benchmark_reader.py for extracting data from the dataset
dev = Benchmark()
files = select_files('en/dev')
dev.fill_benchmark(files)

dev_triples_texts = []  

for i in range(dev.entry_count()):
    #print('====')
    entry = dev.entries[i]
    length = entry.size
    #print(type(length))
    if length == '1':
        category = entry.category
        data = entry.list_triples()
        texts = entry.lexs
        triple = []
        text = []

        for RDF in range(len(entry.list_triples())):
            input_string = data[RDF]
            #dividing in the Triples into 3 strings (Subject, Predicate and Object)
            strings = input_string.split(' | ')
            string1 = strings[0]
            string2 = strings[1]
            string3 = strings[2]

            # Removing unwanted symbols
            if '_' in string1:
                string1 = string1.replace('_', ' ')

            if '_' in string2:
                string2 = string2.replace('_', ' ')

            if '"' in string3:
                string3 = string3.replace('"', '')

            MR = "__subject__ " + string1 + " __predicate__ " + string2 + " __object__ " + string3
            triple.append(MR)

        for lex in range(len(texts)):
            text.append(texts[lex].lex)

        dev_triples_texts.append((triple, [text[0]]))

val_data = dev_triples_texts

# Parse the XML file
tree = ET.parse('en/test/rdf-to-text-generation-test-data-without-refs-en.xml')
root = tree.getroot()

mtriples = []
test_triples_texts = [] 
predicate_test = []
# Iterate over each 'entry' element
for entry in root.findall('.//entry'):
    triple = []
    #print(len(entry.findall('.//modifiedtripleset/mtriple')))
    # Iterate over each 'mtriple' element within 'modifiedtripleset'
    if len(entry.findall('.//modifiedtripleset/mtriple')) == 1:
        for mtriple in entry.findall('.//modifiedtripleset/mtriple'):
            input_string = mtriple.text

            strings = input_string.split(' | ')
            string1 = strings[0]
            string2 = strings[1]
            string3 = strings[2]

            # Removing unwanted symbols
            if '_' in string1:
                string1 = string1.replace('_', ' ')

            if '_' in string2:
                string2 = string2.replace('_', ' ')

            if '"' in string3:
                string3 = string3.replace('"', '')

            MR = "__subject__ " + string1 + " __predicate__ " + string2 + " __object__ " + string3
            predicate_test.append(string2)
            triple.append(MR)
            #mtriples.append(triple)
        
        test_triples_texts.append(triple)

with open('inputTest.txt', 'w', encoding="utf-8") as f:
    i=0
    for triple in test_triples_texts:
        f.write("Triple: " + str(triple) + "\n" + "predicate: "+ predicate_test[i]+"\n\n")
        i=i+1
        
test_data = test_triples_texts

def add_special_tokens(input_ids, output_ids, max_length_diff):
    if output_ids is None:
        return [], [], [], []

    input_ids = [tokenizer.cls_token_id] + input_ids + [tokenizer.sep_token_id]
    attention_mask = [1] * len(input_ids)

    length_diff = len(input_ids) - len(output_ids)

    if length_diff > 0:
        output_ids = output_ids + [tokenizer.pad_token_id] * length_diff
        output_attention_mask = [0] * len(output_ids)
    else:
        output_attention_mask = [0] * len(output_ids)

    return input_ids, attention_mask, output_ids, output_attention_mask



def train_preprocess_data(dataset):
    input_ids_list = []
    attention_mask_list = []
    output_ids_list = []
    output_attention_mask_list = []

    for item in dataset:
        rdf = str(item[0])  # Convert RDF to string
        text = str(item[1])  # Convert text to string

        # Tokenize the RDF and text
        rdf_tokens = tokenizer.tokenize(rdf)
        text_tokens = tokenizer.tokenize(text)

        # Convert tokens to input IDs and output IDs
        input_ids = tokenizer.convert_tokens_to_ids(rdf_tokens)
        output_ids = tokenizer.convert_tokens_to_ids(text_tokens)

        input_ids_list.append(torch.tensor(input_ids))
        attention_mask_list.append(torch.tensor([1] * len(input_ids)))
        output_ids_list.append(torch.tensor(output_ids))
        output_attention_mask_list.append(torch.tensor([1] * len(output_ids)))

    # Convert the input and output data to tensors
    input_ids = pad_sequence(input_ids_list, batch_first=True, padding_value=tokenizer.pad_token_id)
    attention_mask = pad_sequence(attention_mask_list, batch_first=True, padding_value=0)
    output_ids = pad_sequence(output_ids_list, batch_first=True, padding_value=tokenizer.pad_token_id)
    output_attention_mask = pad_sequence(output_attention_mask_list, batch_first=True, padding_value=0)

    return input_ids, attention_mask, output_ids, output_attention_mask




def test_preprocess_data(dataset):
    tokenized_inputs = []

    max_length = 0

    for triple in dataset:
        input_string = triple[0]  # Get the triple string
        input_tokens = tokenizer.encode_plus(
            input_string,  # Triple
            padding='longest',
            truncation=True,
            max_length=512,
            return_tensors='pt'
        )
        input_length = input_tokens['input_ids'].size(1)

        tokenized_inputs.append(input_tokens['input_ids'].squeeze())

        max_length = max(max_length, input_length)

    input_ids = pad_sequence(tokenized_inputs, batch_first=True, padding_value=tokenizer.pad_token_id)
    attention_mask = (input_ids != tokenizer.pad_token_id).float()

    return input_ids, attention_mask



with open('output.txt', 'w', encoding="utf-8") as f:
    #f.write(str(train_data))
    i = 0
    for triple, text in triples_texts:
        f.write("Triple: " + str(triple) + "\n")
        f.write("Text: " + str(text) + "\n")
        f.write("Predicate: " + predicate_train[i] + "\n\n")
        i=i+1
        
with open('Dev_output.txt', 'w', encoding="utf-8") as f:
    for triple, text in dev_triples_texts:
        f.write("Triple: " + str(triple) + "\n")
        f.write("Text: " + str(text) + "\n\n")


print(max_triples)

train_input_ids, train_attention_mask, train_output_ids, train_output_attention_mask = train_preprocess_data(train_data)
val_input_ids, val_attention_mask, val_output_ids, val_output_attention_mask = train_preprocess_data(val_data)
test_input_ids, test_attention_mask = test_preprocess_data(test_data)


print("Training set size:", len(train_input_ids))
print("Validation set size:", len(val_input_ids))
print("Test set size:", len(test_input_ids))


model_config = model.config


print(model_config)
```

### Training Model

```python
import torch
import torch.optim as optim
from torch.nn.utils import clip_grad_norm_
from torch.utils.data import DataLoader, TensorDataset

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # Define the device

num_epochs = 10
batch_size = 8

train_dataset = TensorDataset(train_input_ids, train_attention_mask, train_output_ids, train_output_attention_mask)
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)


model.train()
model.to(device)
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)
#batchNum = 0
print('1')
for epoch in range(num_epochs):
    total_loss = 0.0
    batchNum = 0
    print('2')
    for batch in train_dataloader:
        batchNum = batchNum+1
        print('Batch: ',batchNum)
        print('3')
        input_ids, attention_mask, output_ids, output_attention_mask = [t.to(device) for t in batch]
        print('4')
        optimizer.zero_grad()
        print('5')
        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            decoder_input_ids=output_ids[:, :-1].contiguous(),
            decoder_attention_mask=output_attention_mask[:, :-1].contiguous(),
            labels=output_ids[:, 1:].contiguous(),
        )
        print('6')

        loss = outputs.loss / batch_size
        loss.backward()
        clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        print('7')
        total_loss += loss.item()
    print('8')
    avg_loss = total_loss / len(train_dataloader)
    print(f"Epoch {epoch+1}/{num_epochs} - Average Loss: {avg_loss}")

# Save the trained model
model.save_pretrained("trained_model")
tokenizer.save_pretrained("tokenizer")
```

```python
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration

train_matching_triple = []
train_matching_text = []
train_matching_index = []

# Rest of your code...

# Inside the loop where you find the matching triples
for x in range(len(predicate_test)):
    test_pred = predicate_test[x]
    for i in range(len(predicate_train)):
        if test_pred == predicate_train[i]:
            train_matching_index.append(i)
            train_matching_triple.append(train_data[i][0])
            train_matching_text.append(train_data[i][1][0])
            print("Test RDF: ", test_data[x])
            print("Test predicate: ", test_pred)
            print("Matching Triple:", train_data[i][0])
            print("Matching Text:", train_data[i][1][0])

            # Generate the edited text using the T5 model
            model_name = 't5-base'
            tokenizer = T5Tokenizer.from_pretrained(model_name)
            model = T5ForConditionalGeneration.from_pretrained(model_name)

            prompt = train_data[i][1][0]
            input_text = prompt + " <mask>"
            input_ids = tokenizer.encode(input_text, return_tensors='pt')
            output = model.generate(input_ids)
            edited_text = tokenizer.decode(output[0], skip_special_tokens=True)

            print("Edited Text:", edited_text)

            break  # Stop iterating once a match is found

# Rest of your code...

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

with open('generated_text.txt', 'w', encoding="utf-8") as f:
    model.eval()
    model.to(device)
    test_input_ids = test_input_ids.to(device)
    test_attention_mask = test_attention_mask.to(device)

    generated_ids = model.generate(input_ids=test_input_ids, attention_mask=test_attention_mask)

    for i in range(len(generated_ids)):
        print('heko')
        generated_text = generated_ids[i]
        f.write("Generated Text: " + generated_text + "\n")

        if i < len(train_matching_triple):
            if train_matching_triple[i] is not None:
                # Tokenize the matching triple from the train dataset
                train_input_tokens = tokenizer.encode_plus(
                    train_matching_triple[i],  # Triple
                    padding='longest',
                    truncation=True,
                    max_length=512,
                    return_tensors='pt'
                )

                # Move the train input tokens to the same device as the model
                train_input_ids = train_input_tokens['input_ids'].to(device)
                train_attention_mask = train_input_tokens['attention_mask'].to(device)

                # Generate similar text using the train input tokens
                train_generated_ids = model.generate(
                    input_ids=train_input_ids,
                    attention_mask=train_attention_mask
                )

                # Decode the generated train text and write it to the file
                train_generated_text = tokenizer.decode(train_generated_ids[0], skip_special_tokens=True)
                f.write("Generated Text (Train): " + train_generated_text + "\n")
            else:
                f.write("No matching triple found in the train dataset.\n")
        else:
            f.write("No matching triple found for the generated text.\n")
```
